```python
from datetime import datetime

df_results = pd.read_csv('data/Results_201210.csv', delimiter=';')
df_results['Timestamp'] = df_results['DATE'] + ' ' + df_results['HEURE']
df_results['Timestamp'] = df_results['Timestamp'].apply(lambda x : datetime.strptime(x, '%d/%m/%Y %H:%M:%S'))
df_results['Day'] = df_results['Timestamp'].apply(lambda x : x.day)
df_results['date_rev'] = df_results['Timestamp'].apply(lambda x : str(x.year)+ '{:02d}'.format(x.month)+ '{:02d}'.format(x.day))
df_results['time'] = df_results['Timestamp'].apply(lambda x : '{:02d}'.format(x.hour)+ '{:02d}'.format(x.minute)+ '{:02d}'.format(x.second))

df_results.head()
```

```python
filtered_df = df_results.loc[df_results['Timestamp'].dt.month == 10]
filtered_df
```

```python
def getData(file):
    with open(file, 'r+') as f:
        next(f)
        line = f.readline()
        line = line.replace('\n','').split(';')
    return line[-5:]
```

```python
#---------------------------- prepare the data columns to add
sensor_1 = []
sensor_2 = []
sensor_3 = []
sensor_4 = []

for index, row in filtered_df.iterrows():
    #---------------------------- look for all the files that contains data of the item with Serial_Number N and on the Date D
    #print(SerialN, row['DATE'], row['HEURE'], row['RESULT'], row['time'])
    files = [f for f in files_list if row['SERIAL_NUMBER'] in f and row['date_rev'] in f]
    #print(files)
    
    #---------------------------- filter the files to find the file that contains data at the exact time and get the data
    if len(files) == 0:
        sensor_1.append(None)
        sensor_2.append(None)
        sensor_3.append(None)
        sensor_4.append(None)
    
    if len(files) == 1:
        data = getData(files[0])
        sensor_1.append(data[0])
        sensor_2.append(data[1])
        sensor_3.append(data[3])
        sensor_4.append(data[4])
        
    if len(files) > 1:
        found =False
        i=0
        f=''
        while (found==False and i< len(files)):
            if int(files[i].split('_')[2]) > int(row['time']):
                found=True
                f=files[i]
            i=i+1
        data = getData(f)
        sensor_1.append(data[0])
        sensor_2.append(data[1])
        sensor_3.append(data[3])
        sensor_4.append(data[4])
    
    #----------------------------- 

filtered_df['sensor_1'] = sensor_1
filtered_df['sensor_2'] = sensor_2
filtered_df['sensor_3'] = sensor_3
filtered_df['sensor_4'] = sensor_4

filtered_df.head()
```

```python
filtered_df["RESULT"] = np.where(filtered_df["RESULT"] == "G", 0, 1)
filtered_df
```
